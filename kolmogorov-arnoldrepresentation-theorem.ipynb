```python
import numpy as np
import matplotlib.pyplot as plt

# Define the complex function Francesca
def francesca(x, y):
    return np.sin(x) * np.exp(-y**2) + np.cos(x * y) * np.sqrt(np.abs(x))

# Define the univariate functions (cosine, exponential, and square root)
def cosine_func(x):
    return np.cos(x)

def exponential_func(y):
    return np.exp(-y**2)

def square_root_func(x):
    return np.sqrt(np.abs(x))

# Approximate Francesca using the univariate functions
def francesca_approximation(x, y):
    return cosine_func(x) * exponential_func(y) + cosine_func(x * y) * square_root_func(x)

# Generate sample points
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)

# Evaluate Francesca and her approximation
Z_francesca = francesca(X, Y)
Z_approximation = francesca_approximation(X, Y)

# Plot the results
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))

ax1.contourf(X, Y, Z_francesca, cmap='viridis')
ax1.set_title("Francesca: The Complex Function")
ax1.set_xlabel("x")
ax1.set_ylabel("y")

ax2.contourf(X, Y, Z_approximation, cmap='viridis')
ax2.set_title("Francesca's Approximation: A Dance of Univariate Functions")
ax2.set_xlabel("x")
ax2.set_ylabel("y")

plt.tight_layout()
plt.show()
```
